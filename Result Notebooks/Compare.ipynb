```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
aes_df = pd.read_csv('../Results/AES_results.csv')
des_df = pd.read_csv('../Results/DES_results.csv')
rsa_df = pd.read_csv('../Results/RSA_results.csv')
```

```python
df_columns = aes_df.keys()
df_columns
```

```python
plt.figure(figsize=(10, 5))
x = aes_df[df_columns[0]]
plt.plot(x, aes_df[df_columns[1]], label='AES', marker='o')
plt.plot(x, des_df[df_columns[1]], label='DES', marker='o')
plt.plot(x, rsa_df[df_columns[1]], label='RSA', marker='o')
plt.xlabel('Input Word Length')
plt.ylabel('Encryption Time (ms)')
plt.title('Encryption Time for AES, DES and RSA')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
x = aes_df[df_columns[0]]
plt.plot(x, aes_df[df_columns[2]], label='AES', marker='o')
plt.plot(x, des_df[df_columns[2]], label='DES', marker='o')
plt.plot(x, rsa_df[df_columns[2]], label='RSA', marker='o')
plt.xlabel('Input Word Length')
plt.ylabel('Decryption Time (ms)')
plt.title('Decryption Time for AES, DES and RSA')
plt.legend()
plt.show()
```
