```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../Results/AES_results.csv')
df_columns = df.keys()
df_columns
```

```python
df.head(17)
```

## The values marked NaN in Throughput columns mean that they are not defined and mean essentially Infinity.

```python
df.plot(x=df_columns[0], y=df_columns[1:3], kind='line', marker="o")
plt.xticks(rotation=25)
plt.xlabel('Input Word Length')
plt.ylabel('Time (ms)')
plt.title('AES - Input Word Length vs Encryption Time')
plt.grid()
plt.legend()
plt.show()
```
